```python
import pandas as pd
df = pd.read_csv('pokemon-spawns.csv')
df = df[5546:6789]
df
```

```python
import folium
map = folium.Map(
    location = [df['lat'][6787],df['lng'][6787]],
    zoom_start=12,
    tiles = 'openstreetmap'
)

for i in range(len(df)):
    asd = i+5546
    icon_url = f"https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{df['num'][asd]}.png"
    icon = folium.features.CustomIcon(icon_url,icon_size=(50, 50))
    folium.Marker(
    location=[df['lat'][asd],df['lng'][asd]],
    icon = icon,
    tooltip=f"<img src=https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{df['num'][asd]}.png></img"
    ).add_to(map)
map
```
